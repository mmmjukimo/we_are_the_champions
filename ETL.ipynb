```python
import pandas as pd
import numpy as np
import os
```

```python
train = pd.read_csv("data/train.csv")
test = pd.read_csv("data/sample_submission.csv")

train_img = os.listdir("data/train/")
test_img = os.listdir("data/test/")
train_df = pd.DataFrame(train_img, columns=['image_id'])
test_df = pd.DataFrame(test_img, columns=['image_id'])

print("Number of Train CSV : ", len(train))
print("Number of Total Train Image : ", len(train_df))
print("Number of Test CSV : ", len(test))
print("Number of Test Train Image : ", len(test_df))
```

```python
# 把图片名称的颜色和ID分成两栏
color = []
for n in train_img:
    if "red" in n:
       color.append('red')
    elif "blue" in n:
       color.append('blue')
    elif "yellow" in n:
       color.append('yellow')
    elif "green" in n:
       color.append('green')
train_df['c_name'] = pd.DataFrame(color)

color = []
for n in test_img:
    if "red" in n:
       color.append('red')
    elif "blue" in n:
       color.append('blue')
    elif "yellow" in n:
       color.append('yellow')
    elif "green" in n:
       color.append('green')  
test_df['c_name'] = pd.DataFrame(color)

# 提取ID
train_df['id'] = train_df['image_id'].str.split('_').str[0]
test_df['id'] = test_df['image_id'].str.split('_').str[0]

# sort by id, color
train_df = train_df.sort_values(by=['id', 'c_name']).reset_index(drop=True)
test_df = test_df.sort_values(by=['id', 'c_name']).reset_index(drop=True)
```
